```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
# Load image
img = cv.imread('../images/council_house.jpg', 0)
img = img.astype(np.float)
# Check if image is loaded fine
if img is None:
    raise Exception("File not found")
    
# Resize image
height, width = img.shape
img = cv.resize(img, (500 * width // height, 500))

# Show
plt.figure(dpi=150)
plt.imshow(img, cmap='gray')
```

```python
# Calculate the gradient
ddepth = cv.CV_64F
Ix = cv.Sobel(img, ddepth, 1, 0)
Iy = cv.Scharr(img, ddepth, 0, 1)
```

```python
import simple_harris
import numpy as np
from simple_harris import get_window
```

```python
myIx, myIy = simple_harris.mySobel(img)
```

```python
Ixx, Iyy = simple_harris.derivative(img)
```

```python
fig = plt.figure(dpi=300)

ax1 = fig.add_subplot(3, 2, 1)
ax1.imshow(myIx, cmap='gray')

ax2 = fig.add_subplot(3, 2, 2)
ax2.imshow(myIy, cmap='gray')

ax3 = fig.add_subplot(3, 2, 3)
ax3.imshow(Ix, cmap='gray')

ax4 = fig.add_subplot(3, 2, 4)
ax4.imshow(Iy, cmap='gray')

ax5 = fig.add_subplot(3, 2, 5)
ax5.imshow(Ixx, cmap='gray')

ax6 = fig.add_subplot(3, 2, 6)
ax6.imshow(Iyy, cmap='gray')
```
