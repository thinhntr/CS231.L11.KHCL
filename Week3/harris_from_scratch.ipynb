```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import simple_harris
```

```python
# Load image
img = cv2.cvtColor(cv2.imread('../images/council_house.jpg'), cv2.COLOR_BGR2RGB)

# Check if image is loaded fine
if img is None:
    raise Exception("File not found")
    
# Resize image
height, width, _ = img.shape
img = cv2.resize(img, (500 * width // height, 500))

gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY).astype(np.float)
```

```python
# Show
plt.figure(dpi=150)
plt.imshow(img, cmap='gray')
```

```python
response = simple_harris.simple_harris(gray)
```

```python
plt.imshow(response, cmap='gray')
```

```python
copy = np.array(img)
copy[response>0.00001*response.max()]=[0, 0, 255]
plt.imshow(copy)
```
